## Gràfiques de pèrdues per epoch

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('gpt2-losses.csv')
```

```python
plot1 = df.plot.line(x = 'epoch', y = ['training loss', 'validation loss'] , color = ['blue', 'red'])
```

```python
fig1 = plot1.get_figure()
fig1.savefig('gpt2-train-losses.png')
```

## Mètriques per epoch (train-set)

```python
df = pd.read_csv('gpt2-losses.csv')
```

```python
plot2 = df.plot.line(x = 'epoch', y = ['f1', 'roc_auc', 'accuracy'] , color = ['blue', 'red', 'green'])
```

```python
fig2 = plot2.get_figure()
fig2.savefig('gpt2-train-set-metrics.png')
```

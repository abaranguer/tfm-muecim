```python
import json
import os
import numpy as np
```

```python
def getData(raw):
    data = []
    for item in raw:
        itemShape = len(np.shape(item))
        if (itemShape == 1):
            for part in raw:
                data.append(part)
        else:
            data.append(item)

    return data

def getLines(data):
    linesFlattened = []
    for item in data:
        if type(item) == str:
            lines  = item.split('\n')
        else:
            lines = item[0].split('\n')

        for line in lines:
            linesFlattened.append(line)
        
    return linesFlattened;
```

```python
pathJson = './data/datasets/EURLEX57K'
pathText = './datatxt/datasets/EURLEX57K'
folders = ['dev','test','train']

for folder in folders:
    fullPathJson = os.path.join(pathJson,folder)
    fullPathText = os.path.join(pathText,folder)
    print(f'Iterating through folder {folder}')
    
    ix = 0
    for root, _, fileNames in os.walk(fullPathJson):
        for fileName in fileNames:
            fullJsonFileName = os.path.join(fullPathJson, fileName)
            with open(fullJsonFileName, 'r') as fd:
                jsonObj = json.load(fd)
                title = jsonObj.get('title')
                header = jsonObj.get('header')
                recitals = jsonObj.get('recitals')
                mainBody = jsonObj.get('main_body')
                attachments = jsonObj.get('attachments')
                rawFullText = [title, header, recitals, mainBody, attachments]
                rawFlattenedData = getData(rawFullText)

            fileName = fileName + '.txt'
            fullTextFileName = os.path.join(fullPathText, fileName)
            with open(fullTextFileName, 'w') as fdt:
                lines = getLines(rawFlattenedData)
                for line in lines:
                    fdt.write(line + '\n')

            ix += 1

            if (ix % 1000) == 0:
                print(f'Processing file {ix}')

    print('Done!')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bert-50-labels-losses-metrics-validation-set.csv')
```

```python
plot1 = df.plot.line(x = 'Epoch', y = ['Training Loss', 'Validation Loss'] , color = ['blue', 'red'])
```

```python
fig1 = plot1.get_figure()
fig1.savefig('50-labels-BERT-training-losses.png')
```
